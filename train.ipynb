```python
# drug 데이터를 불러옵니다.
import pandas as pd

data = pd.read_csv(r'C:\sw\first_project\pose_data.csv')
data
```

```python
#레이블 선택
data_label = data.iloc[:,0]
data_label
```

```python
# drug 데이터에서 특징 값을 선택합니다.
# 약물 처방에 필요한 데이터를 선택하고, 불필요한 것은 삭제하거나 선택하지 않습니다.
data_data = data.iloc[:,1:]
data_data
```

```python
# 데이터를 학습 데이터와 검증 데이터로 분할합니다.
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_data, data_label, test_size=0.2)
```

```python
# KNN 모델을 생성하여 학습

from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, n_jobs=-1) #n_naighbors = k값 지정
knn.fit(X_train, y_train)
```

```python
#KNN 모델 예측 정확도
from sklearn.metrics import accuracy_score
y_pred = knn.predict(X_test)

print('예측 정확도:%f' %(accuracy_score(y_test,y_pred)))
```

```python
# 모델을 파일로 저장
import joblib

model_filename = 'knn_model.pkl'
joblib.dump(knn, model_filename)

print(f"모델이 '{model_filename}' 파일로 저장되었습니다.")
```
